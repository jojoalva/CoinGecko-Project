# Data Cleaning Exercise

```python
import pandas as pd
import sqlite3
import requests


# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, text, inspect, func
```

```python
#URL to obtain API data from

url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=100&page=1&sparkline=false"

response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    data = response.json()  # Convert response to JSON
else:
    print("Failed to fetch data")
```

```python
df = pd.DataFrame(data)
df
```

```python
df.drop('roi', axis = 1, inplace=True)
```

```python
df.info()
```

```python
column_names = df.columns
column_names
```

```python
df.to_csv('Data in CSV File/coindata.csv')
```

```python
conn = sqlite3.connect('coins.sqlite')
```

```python
df.to_sql("coindata", conn, if_exists="replace")
```

```python
conn.commit()
```

```python
conn.close()
```

```python
# Create the engine with your database connection string
engine = create_engine("sqlite:///coins.sqlite")

# Establish a connection
connection = engine.connect()

# Execute the SQL query
query = text('SELECT * FROM coindata')
result = connection.execute(query).fetchall()

# Display the results
print(result)

# Close the connection
connection.close()
```

```python
# Now, you can work with 'result'
for row in result:
    print(row)
```

```python
inspector = inspect(engine)
tables = inspector.get_table_names()
tables
```
